```python
from datasets import load_dataset

df = load_dataset("SebastianBodza/LeetCode_UserSubmissions", split='train').to_pandas()
df
```

```python
df.drop(columns=['id', 'title', 'commentCount', 'topLevelCommentCount', 'pinned', 'isFavorite', 'solutionTags', 'post', 'searchMeta', 'Java', 'Cpp', 'Python3', 'Python', 'C', 'JavaScript', 'Golang', 'Csharp', 'Typescript', 'Swift', 'Kotlin', 'Rust', 'Ruby', 'Php', 'Scala', 'Dart', 'Elixir', 'Racket', 'Erlang', 'Mysql', 'Mssql', 'Pythonml', 'detailedViewCount'], inplace=True)
df
```

```python
import re

def remove_urls(text):
    url_pattern = re.compile(r'(https?|ftp|file)://[-A-Za-z0-9+&@#/%?=~_|!:,.;]+[-A-Za-z0-9+&@#/%=~_|]')
    return url_pattern.sub(r'', text)

df['content'] = df['content'].apply(remove_urls)
df
```

```python
df.dropna(inplace=True)
df
```

```python
df.rename(columns={'viewCount': 'view_count', 'titleSlug': 'title_slug', 'voteCount': 'vote_count'}, inplace=True)
df
```

```python
df_leetcode = load_dataset('kreimben/leetcode_questions', split='train').to_pandas()
df_leetcode = df_leetcode.drop(columns=['id', 'title', 'similar_question_ids', 'success_rate', 'total_submission', 'total_accepted', 'question_likes', 'question_dislikes',])
df_leetcode
```

```python
df = df_leetcode.merge(df, how='inner', on='title_slug')
df
```

```python
datasets_id = 'kreimben/leetcode_user_submissions'
token = ''

from datasets import Dataset

dataset = Dataset.from_pandas(df)

dataset.push_to_hub(datasets_id, token=token)
```
